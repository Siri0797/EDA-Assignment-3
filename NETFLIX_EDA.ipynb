```python
most_watched = df.sort_values(by='show_id', ascending=False).head(10)
most_watched
```

```python
# Convert 'rating' column to strings before creating the histogram.
plt.hist(df['rating'].astype(str))
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
average_ratings = df.groupby('type')['rating'].mean()
average_ratings
```

```python
# Assuming 'date_added' is in a suitable format, handling potential leading/trailing spaces
df['release_month'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce').dt.month
monthly_releases = df.groupby('release_month')['title'].count()
monthly_releases
```

```python
# Group by 'genre' and sum the 'view_count'
genre_views = df.groupby('listed_in')['show_id'].count()
genre_views
```

```python
# Group by 'release_year' and count the movies
yearly_releases = df.groupby('release_year')['title'].count()
yearly_releases
```

```python
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')

movies_by_year = df['release_year'].value_counts().sort_index()


plt.figure(figsize=(10,6))
plt.bar(movies_by_year.index, movies_by_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Netflix Movies by Year')
plt.show()
```

```python
# Filter for the show title and select the desired columns
house_of_cards_info = df[df['title'] == 'House of Cards'][['show_id', 'director']]
house_of_cards_info
```

```python
# Filter for movies released in 2000
movies_2000 = df[(df['release_year'] == 2000) & (df['type'] == 'Movie')]
movies_2000
```

```python
# Filter for TV shows released in India and select the 'title' column
indian_tv_shows = df[(df['country'] == 'India') & (df['type'] == 'TV Show')]['title']
indian_tv_shows
```

```python
# Group by 'director' and count the number of shows/movies
director_counts = df.groupby('director')['title'].count().sort_values(ascending=False).head(10)
director_counts
```

```python

tom_cruise_count = df[df['cast'].str.contains('Tom Cruise', na=False)].shape[0]
print(tom_cruise_count)
```

```python
# Filter for movies with "TV-14" rating and released in Canada
tv14_canada_movies = df[(df['rating'] == 'TV-14') & (df['country'] == 'Canada') & (df['type'] == 'Movie')]
tv14_canada_movies
```
